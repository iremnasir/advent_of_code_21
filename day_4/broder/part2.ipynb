```python
f = open("data.txt","r")
input = f.readlines()

numbers = input.pop(0).split(',')
board_amount = len(input) / 6

input.pop(0)

boards = []
i = 0
while i < board_amount:
    boards.append([0, [x.split() for x in input[:5]], 0])
    input = input[6:]
    i += 1
```

```python
def column_check(board):
    k = 0
    while k < 5:
        bingo = tuple(zip(*board))[k].count('x') == 5
        if bingo:
            return bingo
        k += 1
```

```python
def loopy():
    bingo_order = 1
    for number in numbers:
        for i, board in enumerate(boards):
            if boards[i][0] > 0:
                # print(f"Skipping {board}")
                continue
            for j, board_row in enumerate(board[1]):
                boards[i][1][j] = ['x' if item == number else item for item in board_row]
                # print(f"{i} {number} {boards[i][1][j]}")
                if boards[i][1][j].count('x') == 5 or column_check(boards[i][1]):
                    boards[i][0] = bingo_order
                    boards[i][2] = int(number)
                    bingo_order += 1


    last_bingo = max(boards, key=lambda item: item[0])
    # print('Maximum value:', max_value, "At index:",idx, "with number", number)
    print(boards)
    # print(last_bingo)
    return (last_bingo[1], last_bingo[2])                   
```

```python
import itertools
bingo_board, number = loopy()
print(bingo_board, number)

print(sum(list(map(int, filter(lambda x: x != 'x', list(itertools.chain(*bingo_board)))))) * int(number))
# 65457 too high
# 38181 too high
# 27027
```
